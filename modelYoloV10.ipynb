```python
from ultralytics import YOLO
import os
from IPython.display import Image, display
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov10n.pt')
```

```python
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
```

```python
model.train(
    data=r'C:\Users\stanl\OneDrive\Documents\research\dataset\dataset.yaml',
    epochs=50,
    imgsz=640,
    batch=16,   
    name='jar_dent_detection_YOLOV10',
    pretrained=True
)
```

```python
exp_name = 'jar_dent_detection_YOLOV10'
path = f'runs/detect/{exp_name}'
```

```python
display(Image(filename=f'{path}/results.png'))
display(Image(filename=f'{path}/confusion_matrix.png'))
display(Image(filename=f'{path}/F1_curve.png'))
display(Image(filename=f'{path}/PR_curve.png'))

csv_path = f'{path}/results.csv'
df = pd.read_csv(csv_path)

plt.figure(figsize=(12, 6))
plt.plot(df['epoch'], df['metrics/precision(B)'], label='Precision')
plt.plot(df['epoch'], df['metrics/recall(B)'], label='Recall')
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP@0.5')
plt.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP@0.5:0.95')
plt.plot(df['epoch'], df['train/box_loss'], label='Train Box Loss')
plt.plot(df['epoch'], df['train/cls_loss'], label='Train Class Loss')
plt.plot(df['epoch'], df['val/box_loss'], label='Val Box Loss', linestyle='--')

plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Training Metrics Over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
